Ejemplo Práctico de Creación de Características
Contenido:

Dataset de ejemplo: Usando el conjunto de datos 'titanic' de Seaborn.





```python
import seaborn as sns
df = sns.load_dataset('titanic')

df
```

Creación de nuevas características:

Longitud del nombre:



```python
df['longitud_nombre'] = df['who'].str.len()

df['longitud_nombre']
```

Familiares a bordo:



```python
df['familiares_a_bordo'] = df['sibsp'] + df['parch']

df['familiares_a_bordo']
```

Tarifa por persona:



```python
df['tarifa_por_persona'] = df['fare'] / (df['familiares_a_bordo'] + 1)

df['tarifa_por_persona']
```

Visualización del impacto:

Comparar la supervivencia según las nuevas características.


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='familiares_a_bordo', y='survived', data=df)
plt.title('Supervivencia según familiares a bordo')
plt.show()
```
